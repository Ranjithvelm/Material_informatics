### Question

Plot mean square error (MSE) for various values of w_vec (use the code associated with arbitrary model). It will be a 3-D plot with MSE on z-axis, w0_vec on x-axis and w1_vec on y-axis. Does the MSE plot have multiple minimas?  

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

```python
data = pd.read_csv('LR_1D_data.csv')
```

```python
def lr_1d_model(w_cap_vec, x_vec = data['x']):
    w0_cap = w_cap_vec[0]
    w1_cap = w_cap_vec[1]
    
    # Model prediction
    pred = x_vec * w1_cap + w0_cap
    return pred
```

Arbitrary values are taken for w0 and w1.
Here the range of values is from -10 to 10 for both.
And 50 data points has been chosen in between them.

```python
w0_vec = np.linspace(-10,10,50)
w1_vec = np.linspace(-10,10,50)
w0 = []
w1 = []
mse_vec = []
```

```python
for i in w0_vec:
    for j in w1_vec:
        w0.append(i) ## preparing a new array of w0
        w1.append(j) ## preparing a new array of w1
        y_cap = lr_1d_model([i,j])
        err = data['y'] - y_cap
        mse_new = np.mean(np.square(err))
        mse_vec.append(mse_new)
```

### Minimum point in the graph and their corresponding w0 and w1 values are found.

```python
a = int(mse_vec.index(min(mse_vec)))
min_point=[min(mse_vec), w0[a], w1[a]]
print("The minimum MSE and their corresponding w0 and w1")
min_point
```

### 3D plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(w0, w1, mse_vec)
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('mse')
```

### Answer:
Yes, the plot has multiple minimas.
